```python
# Create temporary dataframe to perform transformations on
df_tmp = df
```

```python
# Replace all occurences of 'male' with 1 and 'female' with 0
df_tmp['target'] = df_tmp['target'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
# Replace all occurences of 'male' with 1 and 'female' with 0
df_tmp['sex'] = df_tmp['sex'].apply(lambda x: 1 if x == 'male' else 0)
```

```python
df_tmp.head()
```

```python
# Visualising correlation matrix using a heatmap
corr_matrix = df_tmp.corr()
fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu");
```

```python
# Since there are only a few missing values, we can drop all the rows with missing values
df_tmp.dropna(axis=0, inplace=True)
```

```python
df_tmp.isna().sum()
```

# Model Preparation

```python
# Split data into X and y
X = df_tmp.drop("target", axis=1)
y = df_tmp["target"]
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Different hyperparameter grids for LogisticRegression, KNN and RandomForestClassifier model
log_reg_grid = {"C": np.logspace(-4, 4, 25),
                "solver": ["liblinear"]}


knn_grid = {'n_neighbors': [3, 5, 7, 9],
            'weights': ['uniform', 'distance'],
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
            'p': [1, 2]}

rf_grid = {"n_estimators": np.arange(10, 1000, 50),
           "max_depth": [None, 3, 5, 10],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
# Setup grid hyperparameter search for LogisticRegression
gs_log_reg = GridSearchCV(LogisticRegression(),
                          param_grid=log_reg_grid,
                          cv=5,
                          verbose=True)

# Fit model
gs_log_reg.fit(X_train, y_train)
```

```python
gs_log_reg.best_params_
```

```python
# Setup grid hyperparameter search for KNN
gs_knn = GridSearchCV(KNeighborsClassifier(),
                      param_grid=knn_grid,
                      cv=5,
                      verbose=True)

# Fit model
gs_knn.fit(X_train, y_train)
```

```python
GridSearchCV(cv=5, estimator=KNeighborsClassifier(),
             param_grid={'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
                         'n_neighbors': [3, 5, 7, 9], 'p': [1, 2],
                         'weights': ['uniform', 'distance']},
             verbose=True)
```

```python
gs_knn.best_params_
```

```python
# Setup randomized hyperparameter search for RandomForestClassifier
rs_rf = RandomizedSearchCV(RandomForestClassifier(), 
                           param_distributions=rf_grid,
                           cv=5,
                           n_iter=20,
                           verbose=True)

# Fit model
rs_rf.fit(X_train, y_train)
```

```python
rs_rf.best_params_
```

# Model Evaluations

```python
def eval(model, X_test, y_test):
    y_preds = model.predict(X_test)
    print(classification_report(y_test, y_preds))
    ConfusionMatrixDisplay.from_predictions(y_test, y_preds)
    RocCurveDisplay.from_predictions(y_test, y_preds)
```

```python
eval(gs_log_reg, X_val, y_val)
```

```python
eval(gs_knn, X_val, y_val)
```

```python
eval(rs_rf, X_val, y_val)
```

```python
# Evaluate LogisticRegression model on test data
eval(gs_log_reg, X_test, y_test)
```

```python
# Evaluate RandomForest model on test data
eval(rs_rf, X_test, y_test)
```

```python
# Evaluate RandomForest model on test data
eval(gs_knn, X_test, y_test)
```

```python
# plot histogram for F1 scores of different models
def plot_f1_scores(models, X_test, y_test):
    f1_scores = []

    # F1 score for each model
    for model in models:
        y_preds = model.predict(X_test)
        f1 = f1_score(y_test, y_preds)
        f1_scores.append(f1)

    # Plotting the histogram
    plt.figure(figsize=(10, 6))
    plt.bar(range(len(models)), f1_scores, color=['blue', 'green', 'red'])
    plt.xlabel('Models')
    plt.ylabel('F1 Score')
    plt.title('Comparison of F1 Scores for Different Models')
    plt.xticks(range(len(models)), ['Logistic Regression', 'KNN', 'Random Forest'])
    plt.show()

# List of models to compare
models_to_compare = [gs_log_reg, gs_knn, rs_rf]

# Plot the histogram
plot_f1_scores(models_to_compare, X_test, y_test)
```
